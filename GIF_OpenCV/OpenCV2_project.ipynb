```python
import cv2
import numpy as np
import os

# Mouse callback function to draw or erase on the canvas
def draw(event, x, y, flags, param):
    global drawing, last_point, mode, canvas_temp

    if mode == "freehand":
        if event == cv2.EVENT_LBUTTONDOWN:  # Start drawing
            save_state()  # Save the current state for undo
            drawing = True
            last_point = (x, y)
        elif event == cv2.EVENT_MOUSEMOVE and drawing:  # Continue drawing
            cv2.line(canvas, last_point, (x, y), color, thickness)
            last_point = (x, y)
        elif event == cv2.EVENT_LBUTTONUP:  # Stop drawing
            drawing = False

    elif mode == "line":
        if event == cv2.EVENT_LBUTTONDOWN:  # Start line
            save_state()  # Save the current state for undo
            drawing = True
            last_point = (x, y)
        elif event == cv2.EVENT_MOUSEMOVE and drawing:  # Draw line preview
            canvas_temp = canvas.copy()
            cv2.line(canvas_temp, last_point, (x, y), color, thickness)
        elif event == cv2.EVENT_LBUTTONUP:  # Finalize line
            drawing = False
            cv2.line(canvas, last_point, (x, y), color, thickness)

    elif mode == "circle":
        if event == cv2.EVENT_LBUTTONDOWN:  # Start circle
            save_state()  # Save the current state for undo
            drawing = True
            last_point = (x, y)
        elif event == cv2.EVENT_MOUSEMOVE and drawing:  # Draw circle preview
            canvas_temp = canvas.copy()
            radius = int(((x - last_point[0]) ** 2 + (y - last_point[1]) ** 2) ** 0.5)
            cv2.circle(canvas_temp, last_point, radius, color, thickness)
        elif event == cv2.EVENT_LBUTTONUP:  # Finalize circle
            drawing = False
            radius = int(((x - last_point[0]) ** 2 + (y - last_point[1]) ** 2) ** 0.5)
            cv2.circle(canvas, last_point, radius, color, thickness)

    elif mode == "fill":
        if event == cv2.EVENT_LBUTTONDOWN:  # Fill the clicked region
            save_state()  # Save the current state for undo
            mask = np.zeros((canvas.shape[0] + 2, canvas.shape[1] + 2), np.uint8)
            cv2.floodFill(canvas, mask, (x, y), color)

    elif mode == "eraser":
        if event == cv2.EVENT_LBUTTONDOWN:  # Start erasing
            save_state()  # Save the current state for undo
            drawing = True
            last_point = (x, y)
        elif event == cv2.EVENT_MOUSEMOVE and drawing:  # Continue erasing
            cv2.line(canvas, last_point, (x, y), (255, 255, 255), thickness)
            last_point = (x, y)
        elif event == cv2.EVENT_LBUTTONUP:  # Stop erasing
            drawing = False

# Function to save the current state of the canvas for undo
def save_state():
    undo_stack.append(canvas.copy())
    if len(undo_stack) > max_undo:  # Limit the size of the undo stack
        undo_stack.pop(0)

# Function to undo the last action
def undo():
    global canvas
    if undo_stack:
        canvas[:] = undo_stack.pop()
    else:
        print("No more actions to undo.")

# Function to display names on the canvas
def display_names():
    global canvas
    canvas[:] = np.full((800, 800, 3), 0, dtype=np.uint8)
    cv2.putText(canvas, "@ Innomatics", (130, 100), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 255), 2)
    cv2.putText(canvas, "Anshini Kumbhare", (70, 300), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255,255), 2)
    cv2.putText(canvas, "Sharoni Pathipati", (70, 400), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 255), 2)
    cv2.putText(canvas, "Harshitha Reddy", (70, 500), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 255), 2)
    cv2.putText(canvas, "Ayush Argonda", (70, 600), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 255), 2)

# Initialize variables
drawing = False
last_point = (-1, -1)
color = (0, 0, 0)  # Default color (Black)
thickness = 2  # Default thickness
canvas = np.ones((800, 800, 3), dtype=np.uint8) * 255  # White canvas
canvas_temp = canvas.copy()
mode = "freehand"  # Default mode

# Undo stack
undo_stack = []
max_undo = 20  # Maximum number of undo actions

# Directory to save frames
save_directory = r"C:\Users\ayush\Data Science Python 2024\Machine_learining\Animatons\Gif\gif_outro"
if not os.path.exists(save_directory):  # Create directory if it doesn't exist
    os.makedirs(save_directory)

frame_count = 0  # Counter for naming frames

cv2.namedWindow("Paint Application", cv2.WINDOW_NORMAL)  # Resizable window
cv2.resizeWindow("Paint Application", 800, 800)  # Set initial window size
cv2.setMouseCallback("Paint Application", draw)

# Main loop for drawing and saving frames
while True:
    if mode in ["line", "circle"] and drawing:  # Show preview for line/circle
        cv2.imshow("Paint Application", canvas_temp)
    else:
        cv2.imshow("Paint Application", canvas)

    key = cv2.waitKey(1) & 0xFF
    if key == ord("s"):  # Save frame
        frame_filename = os.path.join(save_directory, f"frame_{frame_count:04d}.png")
        cv2.imwrite(frame_filename, canvas)
        frame_count += 1
        print(f"Frame saved as {frame_filename}")
    elif key == ord("q"):  # Quit application
        break
    elif key == ord("c"):  # Clear canvas
        canvas = np.ones((800, 800, 3), dtype=np.uint8) * 255  # Reset to white
        undo_stack.clear()  # Clear undo history
    elif key == ord("n"):  # Display names
        mode = "names"
        display_names()
        print("Displaying Names")
    elif key == ord("r"):  # Change to red color
        color = (0, 0, 255)
    elif key == ord("b"):  # Change to brown color
        color = (19, 69, 139)  # RGB value for brown
    elif key == ord("k"):  # Change to black color
        color = (0, 0, 0)
    elif key == ord("g"):  # Change to gray color
        color = (128, 128, 128)  # RGB value for gray
    elif key == ord("l"):  # Switch to line drawing mode
        mode = "line"
        print("Switched to Line Drawing Mode")
    elif key == ord("o"):  # Switch to circle drawing mode
        mode = "circle"
        print("Switched to Circle Drawing Mode")
    elif key == ord("h"):  # Switch to freehand drawing mode
        mode = "freehand"
        print("Switched to Freehand Drawing Mode")
    elif key == ord("e"):  # Switch to eraser mode
        mode = "eraser"
        print("Switched to Eraser Mode")
    elif key == ord("u"):  # Undo the last action
        undo()
        print("Undo last action")
    elif key == ord("f"):  # Switch to fill mode
        mode = "fill"
        print("Switched to Fill Mode")

cv2.destroyAllWindows()
```
